```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.patches as patches
# plot one img

source_img_path = '../dataset/images/'

img_source = source_img_path + dataToTxt.image.iloc[0]
img = mpimg.imread(img_source)

numpy_img = np.array(img)
print(numpy_img.shape)


# Create fig and axes

fig = plt.figure(figsize=(12.5,7.5))

ax = fig.add_subplot()
# fig, ax = plt.subplots()



height = numpy_img.shape[0]
width = numpy_img.shape[1]

  

ax.imshow(img, aspect = 'auto')

for i in range(len(dataToTxt)):
    # Create a Rectangle patch, where left is x and top is y
    rect = patches.Rectangle((dataToTxt.left.iloc[i] * width, dataToTxt.top.iloc[i] * height), dataToTxt.width.iloc[i] * width,
                             dataToTxt.height.iloc[i] * height, linewidth=1, edgecolor='r', facecolor='none')

    # Add the patch to the Axes
    ax.add_patch(rect)


plt.show()
```
